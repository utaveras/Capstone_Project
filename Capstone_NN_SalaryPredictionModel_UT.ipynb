# Methods and Data
## Data Preparation
I have 18 columns of input data which includes the target salary column as well. 
## Problem Formulation and Possible Solutions
The problem can be stated as, salary prediction & statistics for the upcoming years from a given year.

> I have two approaches to consider:
> 1. The first approach could be to develop a time series model (such as a recurrent neural net) to predict the statistics and salaries in future time points. However, oftentimes such models require large datasets and are prone to overfitting. Thus, I chose the next approach.<br>
2. The second approach is to train a model that predicts only the statiscs and salary for the next time point (i.e. next season). That way, for example, if I want to make predictions for 10 seasons in a row, I can do that by performing an inference of the salary 10 times, each time taking the statistics and salary for the previous season to predict the next one.


In this project, I decided to implement the second approach using fullly connected neural networks. FCNNs have been successfully used to solve a varaity of problems. My hope is to leverage the expressivness of such models to capture the highly non-linear relation between the features.
> Part of this inspiration came from [3Blue1Brown's video on Neural Networks](https://youtu.be/IHZwWFHWa-w?t=806) where their "simple" image recoigniction model containing 2 hidden layers predicted numbers at 96% accuracy versus a 98% accuracy when one chose to "play with the network" some more. And finally an impressive 99.79% accuracy was achieved by "state of the art" models. Thus, my goal here was to find a 'simple' neural network for the purposes of this project.

```python
display(all_data.describe())
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

all_data_normalized = pd.DataFrame(scaler.fit_transform(all_data.loc[:,:]), columns=all_data.columns,
                                   index=all_data.index)

all_data_save=all_data_normalized.copy()
all_data_normalized

```

```python
# Create Y(Target/Label) DataSet,i.e. 'all_output_normalized', to use in Test/Train Split
# Target = Next Seasons Statistics & Salaries
all_output_normalized = pd.DataFrame(np.ones(all_data_normalized.shape)*np.nan, columns=all_data_normalized.columns,
                                    index=all_data_normalized.index)

for ind, row in all_data_normalized.iterrows():
    ind_next_season = (ind[0], ind[1]+1)
    try:
        all_output_normalized.loc[ind[0], ind[1]] = all_data_normalized.loc[ind_next_season[0], 
                                                                            ind_next_season[1]].values.reshape((1, -1))
    except:
        all_data_normalized.loc[ind[0], ind[1]] = np.nan

all_data_normalized.dropna(axis=0, inplace=True)
all_output_normalized.dropna(axis=0, inplace=True)

display(all_data_normalized.head())
display(all_output_normalized.head())
```

```python
# Sanity Check
print (all_data.shape)
print (all_data_normalized.shape)
salary_std=np.std(all_data["salary"])
salary_mean=np.mean(all_data["salary"])
display(all_data["salary"])
display( np.add((all_data_normalized["salary"]* salary_std), salary_mean)  )
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, LeakyReLU, BatchNormalization
from tensorflow.keras.regularizers import l1,l2
from tensorflow.keras.initializers import TruncatedNormal, glorot_uniform, glorot_normal
from tensorflow.keras import optimizers
```

```python
def short_model():
    model = Sequential()
    model.add(Dense(20, input_dim=18, kernel_initializer='glorot_uniform',activation="relu"))
    model.add(Dense(25, kernel_initializer='glorot_uniform',activation="relu") )
    model.add(Dense(30, kernel_initializer='glorot_uniform',activation="relu"))
    model.add(Dense(18 ) ) 
    optz = optimizers.Adam(learning_rate=0.001)
    model.compile(loss="mean_squared_error", optimizer=optz )
    return model
```

```python
def wide_model():
    model = Sequential()
    model.add(Dense(32, input_dim=18, kernel_initializer='glorot_uniform',activation="relu"))
    model.add(Dense(64, kernel_initializer='glorot_uniform',activation="relu") )
    model.add(Dense(256, kernel_initializer='glorot_uniform',activation="relu"))
    model.add(Dense(18 ))
    optz = optimizers.Adam(learning_rate=0.001)
    model.compile(loss="mean_squared_error", optimizer=optz )
    return model
```

```python
def deep_model():
    model = Sequential()
    model.add(Dense(32, input_dim=18, kernel_initializer='glorot_uniform', activation='relu'))
    model.add(Dense(128, kernel_initializer='glorot_uniform', activation='relu'))
    model.add(Dense(512,  kernel_initializer='glorot_uniform', activation='relu'))
    model.add(Dense(600,  kernel_initializer='glorot_uniform', activation='relu'))
    model.add(Dense(700,  kernel_initializer='glorot_uniform', activation='relu'))
    model.add(Dense(800, kernel_initializer='glorot_uniform', activation='relu'))
    model.add(Dense(1000,  kernel_initializer='glorot_uniform', activation='relu'))
    model.add(Dense(700, kernel_initializer='glorot_uniform', activation='relu'))
    model.add(Dense(400, kernel_initializer='glorot_uniform', activation='relu'))
    model.add(Dense(100,  kernel_initializer='glorot_uniform', activation='relu'))
    model.add(Dense(18 ))
    optz = optimizers.Adam(learning_rate=0.001)
    model.compile(loss="mean_squared_error", optimizer=optz )
    return model
```

```python
def deep_model_unetbased():
    # create model
    model = Sequential()
    model.add(Dense(64, input_dim=18, kernel_initializer='glorot_uniform', activation='relu'))
    model.add(Dense(128, kernel_initializer='glorot_uniform', activation='relu'))

    model.add(Dense(256, kernel_initializer='glorot_uniform', activation='relu'))
    model.add(Dense(512, kernel_initializer='glorot_uniform', activation='relu'))
    model.add(Dense(1024, kernel_initializer='glorot_uniform', activation='relu'))

    model.add(Dense(512, kernel_initializer='glorot_uniform', activation='relu'))
    model.add(Dense(256, kernel_initializer='glorot_uniform', activation='relu'))

    model.add(Dense(128, kernel_initializer='glorot_uniform', activation='relu'))
    model.add(Dense(64, kernel_initializer='glorot_uniform', activation='relu'))
    model.add(Dense(32, kernel_initializer='glorot_uniform', activation='relu'))
    model.add(Dense(18 ))
    # Compile model
    optz = optimizers.Adam(learning_rate=0.001)
    model.compile(loss="mean_squared_error", optimizer=optz )
    return model
```

```python

models=[short_model(), wide_model(), deep_model(), deep_model_unetbased() ]
models_names=["Short Model", "Wide Model", "Deep Model", "Deep Model UnetBased"]
```

## Final Model
>The Short Model does not overfit but it underfits the data (c.f. the [Report/README.md](https://github.com/utaveras/Capstone_Project/blob/master/README.md) for details), whereas the Wide model has a lower loss at a particular point in training before it goes in the wrong direction and overfits...<br><br>
So I will create a compromise between the two models... and present this as the final model
>>Note: regarding optimizers I chose Adam, as it builds upon RMSProp and adds momentum. Plus it is one of the most popular gradient based optimizers. Also, I found out that a learning rate of 0.001 works pretty well for my data.

Furthermore, I reduce the learning rate when the loss value plateaus and reduce the learning rate to give it a chance to exploit the neigboring points with the hope of findding better weights with even lower loss value.

```python
def short_wide_final():
    model = Sequential()

    model.add(Dense(30, kernel_initializer="glorot_uniform" ))
    model.add(LeakyReLU())
    model.add(Dense(40, kernel_initializer="glorot_uniform" ))
    model.add(LeakyReLU())  
    model.add(Dense(50, kernel_initializer="glorot_uniform" ))
    model.add(LeakyReLU())
    model.add(Dense(18))

    optz = optimizers.Adam(learning_rate=0.001)
    model.compile(loss="mean_squared_error", optimizer=optz )
    return model
```

```python
final_model = short_wide_final()
```

```python
# ReduceLROnPlateau allows dynamic learning rate reducing based on some validation measurements.
from tensorflow.keras.callbacks import ReduceLROnPlateau
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, mode ="max",patience=3, min_lr=0.0001)
```

```python

final_model_hist=final_model.fit(X_train,Y_train,batch_size=8, validation_split=0.25, epochs=100  , callbacks=[reduce_lr])
```

```python
print (final_model.evaluate(X_train, Y_train))
print (final_model.evaluate(X_test, Y_test))

plt.plot(final_model_hist.history['loss'])
plt.plot(final_model_hist.history['val_loss'])
plt.title( "Final Model Training - Hybrid Model"+' Loss ')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper right')
plt.show()
```
